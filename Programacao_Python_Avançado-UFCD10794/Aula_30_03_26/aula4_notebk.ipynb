Exemplo Geradores - Números Primos

```python
import itertools

def prime_numbers():
    # Handle the first prime
    yield 2
    prime_cache = [2]  # Cache of primes

    # Loop over positive, odd integers
    for n in itertools.count(3, 2):
        is_prime = True

        # Check to see if any prime number divides n
        for p in prime_cache:
            if n % p == 0:  # p divides n evenly
                is_prime = False
                break

        # Is it prime?
        if is_prime:
            prime_cache.append(n)
            yield n

for p in prime_numbers():
    print(p)
    if p > 100:
        break
```

 Mostrar o tamanho em memória do gerador — a grande vantagem dos geradores é que ocupam pouquíssima memória (tipicamente ~200 bytes), independentemente de quantos valores podem gerar, pois calculam os valores um a um em vez de armazená-los todos de uma vez.

```python
import itertools
import sys
 
squares = (x**2 for x in itertools.count(1))
 
print(sys.getsizeof(squares))
```

Gerador de quadrados infinito usando itertools.count(1), que conta a partir de 1. Para cada número x, ele imprime x² e quando o valor ultrapassa 500, chama .close() para encerrar o gerador.

```python
import itertools

squares = (x**2 for x in itertools.count(1))

for x in squares:
    print(x)

    if x > 500:
        squares.close()
```

Exemplo de uso iteradores - Cartas Baralho - Poker

```python
import itertools

# range(2, 11) gera os números 2 a 10
# Concatena com as cartas especiais: Valete, Dama, Rei e Ás
# Converte tudo para string → ['2', '3', ..., '10', 'J', 'Q', 'K', 'A'] — 13 valores no total

ranks = list(range(2, 11)) + ['J', 'Q', 'K', 'A']
ranks = [str(rank) for rank in ranks]

# print(ranks)
# itertools.product(ranks, suits) faz o produto cartesiano — combina cada rank com cada naipe
# Resultado: 13 ranks × 4 naipes = 52 cartas
# Cada carta é uma tupla, ex: ('A', 'Spades'), ('2', 'Hearts')

suits = ['Hearts', 'Clubs', 'Diamonds', 'Spades']
deck = [card for card in itertools.product(ranks, suits)]

# enumerate retorna o índice + o item
# O +1 faz a contagem começar em 1 em vez de 0
# Imprime todas as 52 cartas numeradas

for (index, card) in enumerate(deck):
 print(1 + index, card)

# `itertools.combinations(deck, 5)` gera todas as combinações de **5 cartas** do baralho de 52, sem repetição e sem importar a ordem
# A fórmula matemática é **C(52,5) = 52! / (5! × 47!)** = **2.598.960 mãos**

hands = [hand for hand in itertools.combinations(deck, 5)]

print(f"The number of 5-card poker hands is {len(hands)}.")
```
