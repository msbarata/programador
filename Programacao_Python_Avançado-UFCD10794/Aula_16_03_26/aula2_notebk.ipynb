Classes e Herança Hierárquica

```python
#definição da classe Veiculo - Classe Super
class Veiculo:
    #definição construtor
    def __init__(self, cor, nr_rodas):
        self.cor = cor
        self.rodas = nr_rodas
        print('Veiculo criado com sucesso')
    #defnição de métodos da classe veiculo
    def mostra_att(self):
        return f'Veiculo de cor {self.cor} de {self.rodas} rodas'
   
#DEfinição da subclasse Carro
class Carro(Veiculo):
    #definação do construtor do carro
    def __init__(self,cor, nr_rodas, velocidade, cilindrada):
        super().__init__(cor,nr_rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
        #defnição de métodos da classe veiculo
    def mostra_att(self):
        return f"""Veiculo de cor {self.cor} de {self.rodas} rodas.
            Trata-se dum carro que vai à velocidade de {self.velocidade} Km/h e {self.cilindrada} cilindrada"""
 
#Definição da subclasse Carro
class Camiao(Carro):
    #definação do construtor do carro
    def __init__(self,cor, nr_rodas, velocidade, cilindrada, carga):
        super().__init__(cor,nr_rodas, velocidade, cilindrada)
        self.carga = carga
        #defnição de métodos da classe veiculo
    def mostra_att(self):
        return f"""Veiculo de cor {self.cor} de {self.rodas} rodas.
            Trata-se dum camião que vai à velocidade de {self.velocidade} Km/h e {self.cilindrada} de cilindrada, com {self.carga} toneladas de carga
            """
 



    
    
```

Polimorfismo - Aplicar a mesma função a objetos que são diferentes

```python
#Aplicação das técnicas de polimorfismo
#Aplicamos a mesma função ComoEuSou() a objetos de tipos diferentes
#Invocando ao método meus-Att() específico de cada objeto

def ComoEuSou(familiar):
    familiar.meus_Att()

pai1 = Pai('castanho', 'ruivo')
mae1 = Mae('azuis', 'louro')
familia1 = [filha1, filho1, pai1, mae1]  
print("\nAplicação da técnica de polimorfismo para mostrar diferentes atributos conforme o tipo de objeto")      

for elemento in familia1:
    print(type(elemento).__name__) #Mostra que tipo de elemento corresponde ao chamado na sua respetiva ordem
    ComoEuSou(elemento)
```

Classes e objetos com métodos - privados X públicos
Programação orientada ao objeto requer a não repetição...

```python
class Pessoa:
    #Construtor
    def __init__(self, nome, idade, nif):
        self.nome = nome
        self.idade = idade
        self._nif = nif      # _nif - atributo passou a estar protegido, só objeto tipo Pessoa e Empregado poderá aceder.
 
    #Métodos da classe gets e sets dos atributos dos objetos do tipo classe Pessoa
    def get_nome(self):
        return self.nome
    def set_nome(self,novo_nome):
        self.nome = novo_nome
        return self.nome
 
    def get_idade(self):
        return self.idade
    def set_idade(self,nova_idade):
        self.idade = nova_idade
        return self.idade
   
    def get_nif(self):
        return self._nif
    def set_nif(self,novo_nif):
        self.nif = novo_nif
        return self._nif
 

    
```

Instanciando a classe com o uso do get para mostrar um conteúdo

```python
p1 = Pessoa('Rafael', 60, 1456375528)
print(p1.get_nif())
print(p1.get_nome())
```

Instanciando a classe com o uso do set para alterar um conteúdo

```python
p1.set_nome('Rafael Moreira')
print(p1.get_nome())
```

O correto é alterar conteúdo acedendo pelo método da classe, no entanto, o Python permite aceder diretamente ao atributo da classe: p1.nome = ''

```python
p1.nome = 'Manuel'
print(p1.get_nome())
```

Restrições do python - classes privadas

```python
class Empregado(Pessoa):
    #construtor da classe Empregado
    def __init__(self, nome, idade, nif, empresa, salario):
        super().__init__(nome, idade, nif)
        self.empresa = empresa
        self.salario = salario
 
    #Métodos do Empregado
    def get_empresa(self):
        return self.empresa
    def set_empresa(self,nova_empresa):
        self.empresa = nova_empresa
        return self.idade
   
    def get_salario(self):
        return self.salario
    def set_salario(self,novo_salario):
        self.salario = novo_salario
        return self.salario
    
```
