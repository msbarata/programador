DEFINIÇÃO DE SUBCLASSE PARA A CLASSE CARRO

```python
#Definição da superclasse - Carro

class Carro:
    def __init__(self, marca, modelo, ano):
          
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.odometro = 0

#definição dos métodos da classe carro
#método que mostra as características do carro

    def get_descricao_carro(self):
       # nome_completo =f"Carro de {self.ano}, da marca {self.marca}, e do modelo {self.modelo}"
       # return nome_completo
       return f"Carro de {self.ano}, da marca {self.marca}, e do modelo {self.modelo}"
    
#método que lê odômetro do carro

    def ler_odometro(self):
        print (f"Este carro tem {self.odometro} km.")
    
#método que incrementa kilometros ao odômetro
    def incrementa_odometro(self, kilometros):
        self.odometro += kilometros

#Defnição da subClasse - Carro elétrico (precisamos respeitar número de parâmetros da superClasse - Pai)
#Por herdar os métodos da classe pai (Carro), pode usá-los diretamente.
#Inicializa os atributos da superClasse - (Pai)

class CarroEletrico(Carro):
#Construtor da subclasse
    def __init__(self,marca,modelo,ano,autonomia,bateria):
    #Atributos herdados da superClasse
        super().__init__(marca,modelo,ano)

#atributos da subclasse CarroEletrico
        self.autonomia = autonomia
        self.bateria = bateria
    def get_autonomia(self):
        return self.autonomia
    def get_bateria(self):
        return self.bateria
```

```python
ce1 = CarroEletrico('Tesla', 'A', 2025, 200, 1000)
ce1.get_descricao_carro()
```

```python
ce1.ler_odometro()
ce1.incrementa_odometro(20)
ce1.ler_odometro()
ce1.get_autonomia()
ce1.get_bateria()
```

##Escrever num ficheiro - necessário declarar o modo gravar ('w') e o encoding ('utf8')

```python
with open('carla.txt', 'w', encoding='utf8') as ficheiro_carla:
    ficheiro_carla.write("olá, meu nome é carla Godinho/n")
    ficheiro_carla.write("Vamos terminar .../n")
    ficheiro_carla.close()
```

"função para acrescentar texto ao ficheiro - appending

```python
# Segunda execução com 'a' - ADICIONA ao final
with open('carla.txt', 'a', encoding='utf8') as f:
    f.write("Segunda linha\n")
# Resultado: arquivo tem "Primeira linha" e "Segunda linha"
```
