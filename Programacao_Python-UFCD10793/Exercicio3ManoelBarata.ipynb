Tuple - Tipo de dados (valores ao contrário das listas não permitem alteração dos valores, após inseridos)
Utiliza-se este tipo de dados para trabalhar e armazenar dados sensíveis que não deseja que sejam alterados.

```python
tuplo1=(1,2,3, 'GIS')
print(tuplo1)
print(tuplo1[2])
```

```python
tuplo2=(1,2,3,'GIS', [6,4,5])
print(tuplo2)
tuplo2[4][0]=3.4
print(tuplo2)
```

Interseção e União - SET {} Conjuntos e usa-se chavetas para delimitar seus dados. Estes não possuem numeros duplicados.

```python
conjunto1 = {3, 10.5, 'Carla', 2.1}
print(conjunto1) #não mostra em ordem, mas sim aleatória
lista1 = ['Carla', 2, 4.5, 2]
print(lista1)
nova_set = set(lista1) #set é uma função do python que transforma uma lista em conjunto, eliminando os valores que se repetem na lista original
print(nova_set)
print(lista1)
```

```python
conjunto2 = {4, 22, 1, 10}
conjunto3 = {1, 10, 5, 5, 3}
print(conjunto2) #mostra conteúdo original do conjunto2
print(conjunto3) #mostra conteúdo original do conjunto 3; conjuntos não suportam números repetidos
```

```python
print(conjunto2.union(conjunto3)) # Realiza a união entre dois conjuntos, mas ainda elimina os números repetidos.
```

```python
print(conjunto2.intersection(conjunto3)) # Realiza a interseção entre dois conjuntos, mostra apenas os números que aparecem nos dois conjuntos, mas ainda elimina os números repetidos.
```

Dicionários, usam chavetas {} - no entanto são diferentes de conjuntos. possui a relação de um objeto antes dos dois pontos. é um conjunto de pares chave-valor. Exemplo: chave e valor - 'nome': 'joão'. chaves num dicionário são únicas, não se referem. {'ufcd': '10793'}

```python
{'nome':'João'}
{'nome'=='paulo'}
```

```python
a = {'instituição': 'IEFP', 'Nome da UFCD': 'Fundamentos de Python', 'UFCD': 10793}
print(type(a)) #mostra o tipo de variável complexa - dicionário
print(len(a)) #mostra a quantidade de campos do dicionário
print (a.keys()) #lista o conteúdo das chaves do dicionário
print (a.values()) #lista o conteúdo dos valores do dicionário
print(a.get("instituição")) #mostra o valor anexado ao campo chave 'instituição'
print(a['Nome da UFCD'])



```

pop e del apagam objetos no dicionário

```python
print(a)
```

```python
a['Nome da UFCD'] = 'Programação Avançada com Python'
print(a)
```

```python
a.setdefault('Nome da UFCD','Programação Avançada com Python')
print(a)
```

```python
a.setdefault('UFCD', 10793) #função a.setdefault permite adicionar novas entradas no dicionário, mas se entrada já existe não é alterada desta forma.
print(a)
```

```python
a['UFCD'] = 10794 #inclusão de uma nova UFCD, utilizando as chaves - (as outras chaves estão com valores default, como se fossem campos de uma base de dados)
print(a)
```
