Trabalhando com o tipo de dados - String; 'Python' é um conjunto de objetos do tipo string; a variável 'a' se torna dinamincamente uma classe do tipo string; len quantifica a sequência de strings; [x] o valor entre colchete determina a posição das diversas strings que serão impressas.

```python
a="Python"
print (type(a)) #imprime o tipo da variável a, neste caso "Python" - string
print (len(a)) #len - quantifica o total de carateres contidos na palavra "python"
print (a[1]) #a[1] retorna o caracter da posição 1 da string "python"
print (a[0:3]) #a[0:3] retorna os caracteres das posições 0 a 2 da string "python", pois o "3" é exclusive - (finaliza sem ser incluída)
print (a[-1]) #a[-1] retorna o último caracter do conjunto de string, com uma contagem do fim para o início, cao iniciar do -1(n), -2(o)...

```

Variáveis - objetos (a, b e c): Operações com strings; soma (concactenação entre strings); duplicação de strings; caixa alta e caixa baixa em string; objeto "d" recebe a concatenação das strings a, b e c; o método replace susbstitui strings e o método split divide a string em pedaços e retorna uma lista.

```python
a="Python"
b="->"
c="Fofinho"
print(a+b+c) #realiza a concatenação do conjunto de strings, contidas nas variáveis a,b,c
print(a*2) #duplica a string presente na variável a
print(a.upper()) #upper coloca em caixa alta o conjunto de strings contidos na variável a
print(a.lower()) #upper coloca em caixa baixa  conjunto de strings contidos na variável a
d=a+b+c #faz a concatenação dos conjunto de strings contidos nas variáveis a,b,c
print(d) #imprime o conjunto de strings concatenados na linha anterior
print(d.find("->")) #imprime a posição numérica da string procurada, no caso, "6"
f=d.replace("->", " é ")#replace substitui trechos de variáveis, no entanto, precisa ser salvo em uma nova viariável, no caso (f)
print(d) #imprime conteúdo da variável "d"
print(f) #imprime conteúdo da variável "f"
print(d.split("->")) #divide a string em pedaços, separadas por vírgulas.
```

Importação da Biblioteca sys; formatação de strings, nomeadamente por meio do uso de % e também do método .format, o qual permite inserir strings em locais determinados das frases por meio de números entre chavetas.


```python
import sys
a = "Tem %i computador com %s instalado" % (1,sys.platform) #importa o método %i e %s da biblioteca sys
b = "Tem %.3f computador com %s instalado" % (1,sys.platform) #%i - inteiro; %s - string; %c -caracter; %d - decimal; %f - float; %e - float com expoente
c = "Tenho {0} {1} com {2} instalado e o binário do executável está em {3}".format (1, "portátil", sys.platform, sys.executable)
print(a)
print(b)
print(c)
```

Variáveis Tipo Listas

```python
minhaLista = [1,2,3,'Python outra vez']
print(minhaLista)
print(minhaLista[3]) #mostrar o elemento da lista na posição 3
print(minhaLista[-1]) #mostra o último elemento da lista (contabiliza do final para o início)
```
