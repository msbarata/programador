```python
MESES = {
    1: {
        'nomeTrim': 'primeiro',
        'meses': [
            {'mes': 1, 'nomeMes': 'janeiro'},
            {'mes': 2, 'nomeMes': 'fevereiro'},
            {'mes': 3, 'nomeMes': 'março'}
        ]
    },
    2: {
        'nomeTrim': 'segundo',
        'meses': [
            {'mes': 4, 'nomeMes': 'abril'},
            {'mes': 5, 'nomeMes': 'maio'},
            {'mes': 6, 'nomeMes': 'junho'}
        ]
    },
    3: {
        'nomeTrim': 'terceiro',
        'meses': [
            {'mes': 7, 'nomeMes': 'julho'},
            {'mes': 8, 'nomeMes': 'agosto'},
            {'mes': 9, 'nomeMes': 'setembro'}
        ]
    },
    4: {
        'nomeTrim': 'quarto',
        'meses': [
            {'mes': 10, 'nomeMes': 'outubro'},
            {'mes': 11, 'nomeMes': 'novembro'},
            {'mes': 12, 'nomeMes': 'dezembro'}
        ]
    }
}

trimestre = int(input("Digite o número do trimestre [1 a 4]: "))

# Usa get() para obter os dados do trimestre
trim_data = MESES.get(trimestre)

# Condição if/elif/else para tratar a escolha
if trim_data:
    # Obtém os nomes dos meses usando get()
    nome_trimestre = trim_data.get('nomeTrim')
    meses = trim_data.get('meses')    
    print("Meses do %s trimestre: %s"%(nome_trimestre, meses))    
   
else:
    print("Opção errada! Escolha um trimestre entre 1 e 4.")
```
